### OUNoise Simulation & Representation 

```python
import random
import numpy as np
import copy
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
mu=0.
theta=0.0003
sigma=0.002
size=(1,2)
random.seed(0)
```

```python
mu = mu * np.ones(size)
state = copy.copy(mu)
x_axis=[]
y_axis=[]
for i in range(1000):
    x = state
    dx = theta * (mu - x) + sigma * np.random.rand(size[0],size[1])
    state = x + dx
    x_axis.append(state[0][0])
    y_axis.append(state[0][1])
    #states.append(state)

fig = plt.figure()
ax = fig.add_subplot(111)
plt.plot(x_axis,y_axis,label="points")
plt.show()
```
